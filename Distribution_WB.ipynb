```python
import csv
import numpy as np
def create_matrix(filename):
    with open(filename) as f:
        data = csv.reader(f)
        body = []
        for idx, row in enumerate(data):
            if idx == 0: # header
                header = row
            else:
                # We know this is brute force but moving on to plots - would like to try mask eventually
                if "?" in row[0] or "?" in row[1] or "?" in row[3] or "?" in row[4] or "?" in row[7]:
                    pass
                else:
                    body.append(row)
    return header, np.array(body)
```

```python
def stuff ():
    file_list = ['./cleveland_heart.csv','./hungarian_heart.csv', './switzerland_heart.csv', './long_beach.csv']
    data_matrices, location_names = [], []
    for f in file_list:
        location_name = f.replace('.csv', '').replace('./', '')
        location_names.append(location_name)
        header, data = create_matrix(f)
        data_matrices.append({ header[i]: data[:,i] for i in range(len(header))})
    location_data = dict(zip(location_names, data_matrices))
    return location_data

location_data = stuff()
```

```python
import matplotlib.pyplot as plt

def plot_ages(plot_pt, location_data):
    fig, axes = plt.subplots(2,2, figsize=(12,12))
    vals = [axes[0][0],axes[0][1],axes[1][0],axes[1][1]]
    idx = 0
    for location, data in location_data.iteritems():
        ages = (data[plot_pt]).astype(np.float)
        vals[idx].hist(ages, bins=(max(ages)-min(ages)))
        vals[idx].set_title(location)
        vals[idx].set_ylabel("Count")
        vals[idx].set_xlabel("Age")
        vals[idx].set_xlim([25, 80])
        idx += 1

plot_ages('age', location_data)
```

```python
def plot_heart_cholesterol(plot_pts, location_data):
    sex, ch, hr = plot_pts
    fig, axes = plt.subplots(2,2)
    axes = plt.axes()
    f_chol_list, m_chol_list, f_hr_list, m_hr_list = [],[],[],[]
    
    for location, data in location_data.iteritems():    
        # Build sub matrix
        sub_matrix = np.zeros(shape=(len(data[sex]),3))
        sub_matrix[:,0] = data[sex]
        sub_matrix[:,1] = data[ch]
        sub_matrix[:,2] = data[hr]
        print location
        
        # Split out data
        female_data = sub_matrix[sub_matrix[:,0] == 0]
        male_data = sub_matrix[sub_matrix[:,0] == 1]
    
        f_chol_list.append((female_data[:,1]).astype(np.float))
        m_chol_list.append((male_data[:,1]).astype(np.float))
        
        f_hr_list.append((female_data[:,2]).astype(np.float))
        m_hr_list.append((male_data[:,2]).astype(np.float))

    axes[0][0] = boxplot(f_chol_list, vert=1)
    axes[1] = boxplot(m_chol_list, vert=1)
    axes[2] = boxplot(f_hr_list, vert=1)
    axes[3] = boxplot(m_hr_list, vert=1)

plot_heart_cholesterol(['sex','cholesterol_mg/dl','max_heart_rate'], location_data)
```
